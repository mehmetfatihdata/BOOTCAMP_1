## 1.Soru

Bu soruda, merkezi limit teoreminin çalışıp çalışmadığını test edeceksiniz. İki adet normal dağılıma sahip 1000 gözlem içeren değişken üreteceksiniz. Bu değişkenlerin, ortalama ve standart sapmasını kendiniz belirleyebilirsiniz. Yapmanız gereken, ilk değişkeni 50 kere üretin ve her seferinde bu değişkenin ortalamasını alın. İkinci değişkeni ise 1000 kere üretin ve yine her seferinde bu değişkenin ortalamasını alın. Daha sonra, iki değişkenin hesapladığınız ortalamalarının histogramını çizdirin. Hangi değişkenin ortalamalarının dağılımı normal dağılıma daha yakın? Sizce Merkezi Limit Theoremi çalışmış görünüyor mu?

```python
örneklem_1 = []
örneklem_ortalaması_1 = []
for tekrar_sayısı in range(1,1000):
    degısken_1 = np.random.normal(5, 35, 1000)
    örneklem_1.append(degısken_1)
    örneklem_ortalaması_1.append(np.mean(örneklem_1))       
örneklem_ortalaması_2 = []    
örneklem_2 = []
for tekrar_sayısı in range(1,50):
    degısken_2 = np.random.normal(5, 35, 50)
    örneklem_2.append(degısken_2)
    örneklem_ortalaması_2.append(np.mean(örneklem_2))
```

```python
baslik_font = {'family': 'arial', 'color': 'darkred','weight': 'bold','size': 13 }
eksen_font  = {'family': 'arial', 'color': 'darkblue','weight': 'bold','size': 10 }
plt.figure(figsize=(15,3), dpi=100)

plt.subplot(121)
plt.plot(örneklem_ortalaması_1,c="y")
plt.hlines(5, 0,1000, colors= 'r' )
plt.plot(örneklem_ortalaması_1[0:1000])
plt.ylim((1,10))
plt.title("Örneklem Sayısı = {}".format(1000), fontdict = baslik_font)
plt.xlabel("Örneklem Sayısı", fontdict = eksen_font)
plt.ylabel("Örneklem Ortalaması", fontdict = eksen_font)

plt.subplot(122)
plt.plot(örneklem_ortalaması_2,c="g")
plt.hlines(5, 0,50, colors= 'r' )
plt.plot(örneklem_ortalaması_2)
plt.ylim((1,10))
plt.title("Örneklem Sayısı = {}".format(50), fontdict = baslik_font)
plt.xlabel("Örneklem Sayısı", fontdict = eksen_font)
plt.ylabel("Örneklem Ortalaması", fontdict = eksen_font)
```

Yüksek standart sapmaya sahip bir popülasyondan yeterli sayıda örneklem seçilmediği takdride veri analiz doğru olmayacaktır. 

## 2.SORU

İki madeni paramızın olduğunu varsayalım. Bunlardan bir tanesi adil (yani, yeteri kadar deneme yapıldığında yazı veya tura 1/2 olasılığına yaklaşmakta), diğeri ise adil değil ve atıldığında 3/4 olasılığında tura elde edilmektedir. Herhangi biri seçilip, n kere deneme yapıldığında, Büyük Sayılar Yasası uzun vadede gelecek tura sayısını tahmin etmemize izin verir mi? Çok sayıda deneme yapıldığında, hangi paranın seçildiğini bilebilir miyiz?

```python
adil=[]
adil_ortalama=[]
for atıs_sayisi in range (1,1000):
    atış=np.random.binomial (1,0.5)
    adil.append(atış)
    adil_ortalama.append(np.mean(atış))
adil_olmayan=[]
adil_olmayan_ortalama=[]
for atıs_sayisi2 in range (1,1000):
    atış2=np.random.binomial (1,0.75)
    adil_olmayan.append(atış2)
    adil_olmayan_ortalama.append(np.mean(atış2))
```

```python
plt.figure(figsize=(10,4), dpi=100)
baslik_font = {'family': 'arial', 'color': 'darkred','weight': 'bold','size': 13 }
eksen_font  = {'family': 'arial', 'color': 'darkblue','weight': 'bold','size': 10 }

plt.subplot(121)
plt.plot(adil_ortalama[0:100])
plt.hlines(0.5,0,100,colors="r")
plt.title("Adil Olanın Tura Gelme Olasılığı", fontdict=baslik_font)
plt.xlabel("Adil Atış", fontdict=eksen_font)
plt.ylabel("Adil Atış Ortlaması", fontdict=eksen_font)

plt.subplot(122)
plt.plot(adil_olmayan_ortalama[0:100])
plt.hlines(0.75,0,100,colors="r")
plt.title("Adil Olmayanın Tura Gelme Olasılığı", fontdict=baslik_font)
plt.xlabel("Adil_Olmayan Atış", fontdict=eksen_font)
plt.ylabel("Adil_Olmayan Atış Ortlaması", fontdict=eksen_font)

plt.tight_layout()
plt.show()
```
