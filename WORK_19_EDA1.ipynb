# Q5: 
Metematik ve okuma notlarında birçok eksik değer olduğunu farketmişsinizdir. Eksik değerleri ortalama, medyan ve enterpolasyon ile tamamlayın. Bu tekniklerin notların dağılımı üzerindeki etkisini karşılaştırın.

```python
def nullhesap(df):
    yüzde=df.isnull().sum()*100/df.shape[0]
    adet=df.isnull().sum()
    return pd.DataFrame(list(zip(yüzde,adet)),columns=["%","Nu"])
nullhesap(ALABAMA).T
```

```python
egitim = pd.read_csv("states_all.csv")
```

```python
ALABAMAA=egitim[egitim['STATE'] == "ALABAMA"]

nullhesap(ALABAMAA).T
```

```python
d_mean=ALABAMAA[['AVG_MATH_4_SCORE','AVG_READING_4_SCORE','AVG_MATH_8_SCORE','AVG_READING_8_SCORE']]
for col in d_mean:
       d_mean[col] = d_mean[col].transform(lambda x: x.fillna(x.mean()))
```

```python
d_median=ALABAMAA[['AVG_MATH_4_SCORE','AVG_READING_4_SCORE','AVG_MATH_8_SCORE','AVG_READING_8_SCORE']]
for col in d_median:
       d_median[col] = d_median[col].transform(lambda x: x.fillna(x.median()))
```

```python
d_inter=ALABAMAA[['AVG_MATH_4_SCORE','AVG_READING_4_SCORE','AVG_MATH_8_SCORE','AVG_READING_8_SCORE']]
for col in d_median:
       d_inter[col] = d_inter[col].transform(lambda x: x.interpolate())
```

```python
fig=plt.figure(figsize=(20,200))
f_list=ALABAMAA[['AVG_MATH_4_SCORE','AVG_READING_4_SCORE','AVG_MATH_8_SCORE','AVG_READING_8_SCORE']]
for i, col in enumerate(f_list):
    df=pd.concat([d_mean[col],d_median[col],d_inter[col]], axis=1)
    
    plt.subplot(len(f_list),4,i+1)
    plt.bar(["FİLL_MEAN","FİLL_MEDİAN", "FİLL_İNTERPOLETA"],df.mean(),color=["r","b","green"], alpha=0.7)
    plt.xticks(rotation="vertical")
    plt.title("{}".format(col))
    
plt.tight_layout()
plt.show()
```
